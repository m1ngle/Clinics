```python
# concatenating lat and long to create a consolidated location as accepted by havesine function
loc['coor'] = list(zip(loc.lat, loc.long))
```

```python
# defining a  function to calculate distance between two locations 
# loc1= location of an existing hotel
# loc2= location of customers

def distance_from(loc1,loc2): 
    dist=hs.haversine(loc1,loc2,unit=Unit.MILES)
    return round(dist,2)
```

```python
# running a loop which will parse customers location one by one to distance from function 
for _,row in loc.iterrows():
    loc[row.Clinic]=loc['coor'].apply(lambda x: distance_from(row.coor,x))
```

```python
loc.replace(0,np.nan, inplace=True)
```

```python
def create_map(loc,clinic_list):
    m = folium.Map(location=[loc.lat.mean(), loc.long.mean()], zoom_start=4, 
                  tiles='CartoDB positron')
    
   
    
    for _, row in loc.iterrows():
         
        if row['Condition']=='Out':
            cluster_colour='red'
        else:
            cluster_colour = 'green'
        
        folium.CircleMarker(
            location= [row['lat'],row['long']],
            radius=5,
            popup= row[['Condition','address']],
            color=cluster_colour,
            fill=True,
            fill_color=cluster_colour
            ).add_to(m)
    
    for _, row in loc.iterrows(): 
        folium.Marker(
            location= [row['lat'],row['long']],
            radius=5,
            popup= row['Clinic'],
            color='red',
            fill=True,
            fill_color='red'
             ).add_to(m)
    #for _, row in loc.iterrows(): 
    #    folium.Circle(
    #        location= [row['lat'],row['long']],
    #        radius=160934,
    #        popup= row['Clinic'],
    #        color='black',
    #        fill=False,
    #    ).add_to(m)
        
    return m
```
